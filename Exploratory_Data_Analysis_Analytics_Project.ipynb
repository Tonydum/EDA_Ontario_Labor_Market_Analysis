```python
import pandas as pd

# Load the dataset
ontario_data = pd.read_csv(r'C:\Data Analytics Project/Ontario_Cleaned_Data.csv')

# Display the first few rows of the dataset
ontario_data_head = ontario_data.head()

# Get basic statistics for numerical columns
ontario_data_description = ontario_data.describe()

ontario_data_head, ontario_data_description
```

Columns:
    
GEOGRAPHY: Refers to the region in Ontario.
    
Labor Force Characteristic: Indicates the characteristic being measured, such as the labor force.
    
Occupation: Lists the specific occupation or category.
    
Age Group: Specifies the age group being considered.
    
Both sexes: Labor force count for both males and females combined.
    
Male: Labor force count for males.
    
Female: Labor force count for females.
    
Year: Year of the data entry.
    
Month: Month of the data entry.
    
Aggregate_Flag: Indicates if data is aggregated.
    
Both sexes Missing: Binary flag indicating if data for both sexes is missing.
    
Male Missing: Binary flag indicating if data for males is missing.
    
Female Missing: Binary flag indicating if data for females is missing.

Summary Statistics:
    
The dataset contains data from the year 2001 to 2020.
The average count of the labor force (both sexes) is around 31.65 (thousands), with a maximum of 8011.4 and a minimum of 0.
The columns 'Both sexes Missing', 'Male Missing', and 'Female Missing' are binary flags (0 or 1) indicating missing data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting style for plots
sns.set_style("whitegrid")

# Initialize the figure
plt.figure(figsize=(20, 15))

# Distribution of labor force (both sexes combined) across years
plt.subplot(2, 1, 1)
sns.lineplot(data=ontario_data, x="Year", y="Both sexes", estimator="sum", ci=None)
plt.title('Total Labor Force (Both Sexes) Across Years')
plt.xlabel('Year')
plt.ylabel('Total Labor Force (in thousands)')

# Distribution of labor force (both sexes combined) across months
plt.subplot(2, 1, 2)
sns.lineplot(data=ontario_data, x="Month", y="Both sexes", estimator="sum", ci=None)
plt.title('Total Labor Force (Both Sexes) Across Months')
plt.xlabel('Month')
plt.ylabel('Total Labor Force (in thousands)')

# Display the plots
plt.tight_layout()
plt.show()
```

The visualizations provide the following insights on the distribution of the labor force across years and months:

Total Labor Force Across Years:

The labor force has shown a general increasing trend from 2001 to 2020.
There are some fluctuations, but overall, the labor market seems to be growing over the years.
Total Labor Force Across Months:

While the monthly distribution shows fluctuations, there isn't a clear seasonal pattern discernible from the data.
The labor force seems to be relatively consistent throughout the months, with some peaks and troughs.

Next, the distribution of the labor force across different age groups and occupations.

```python
# Initialize the figure
plt.figure(figsize=(20, 15))

# Distribution of labor force across age groups
plt.subplot(2, 1, 1)
sns.barplot(data=ontario_data, x="Age Group", y="Both sexes", estimator=sum, ci=None)
plt.title('Distribution of Labor Force Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Total Labor Force (in thousands)')
plt.xticks(rotation=45)

# Distribution of labor force across occupations (Top 10 for visualization purposes)
top_occupations = ontario_data.groupby('Occupation').sum()['Both sexes'].nlargest(10).index
subset_data = ontario_data[ontario_data['Occupation'].isin(top_occupations)]
plt.subplot(2, 1, 2)
sns.barplot(data=subset_data, x="Occupation", y="Both sexes", estimator=sum, ci=None)
plt.title('Distribution of Labor Force Across Occupations (Top 10)')
plt.xlabel('Occupation')
plt.ylabel('Total Labor Force (in thousands)')
plt.xticks(rotation=45)

# Display the plots
plt.tight_layout()
plt.show()
```

The visualizations provide insights on the distribution of the labor force across different age groups and occupations:

Distribution Across Age Groups:

The labor force is predominantly concentrated in the age group of 25-54 years. This is expected as this age bracket typically represents the working age population.
There is a significant drop in the labor force for the age groups beyond 55 years, which can be attributed to retirement and other factors.

Distribution Across Occupations (Top 10):

A few occupations dominate the labor market. It's essential to note that for visualization clarity, we've displayed only the top 10 occupations based on the labor force count.
The occupation categories "All occupations," "Total, all industries," and "Goods-producing sector" are broad categories, so their counts are naturally higher. More specific occupational categories like 
"Retail trade" and "Health care and social assistance" also have significant representation in the labor market.

Missing Data Analysis:
    
a. Calculate the percentage of missing values for each column:
We'll determine which columns have missing data and the extent of the missingness.

b. Visualize the distribution of missing data across years and months:
We'll explore if there are specific periods (years/months) where data tends to be missing more often, which can provide insights into potential patterns or systematic issues.

Let's start by calculating the percentage of missing values for each column.

```python
# Calculate the percentage of missing values for each column
missing_data_percentage = (ontario_data.isnull().sum() / len(ontario_data)) * 100

# Filter out columns with no missing data
missing_data_percentage = missing_data_percentage[missing_data_percentage > 0]

missing_data_percentage.sort_values(ascending=False)
```

There are no missing values in the primary columns of the dataset. 

This is great as it means we won't have to perform imputations or make decisions related to missing data handling for these columns.

However, the dataset does have columns named "Both sexes Missing," "Male Missing," and "Female Missing," which indicate binary flags (0 or 1) for missing data. We can delve deeper into these columns to understand the distribution of these flags across years and months.

Let's visualize the distribution of these missing data flags across years and months.

```python
# Initialize the figure
plt.figure(figsize=(20, 15))

# Distribution of 'Both sexes Missing' across years
plt.subplot(3, 1, 1)
sns.countplot(data=ontario_data[ontario_data['Both sexes Missing'] == 1], x="Year", color='blue')
plt.title('Distribution of "Both sexes Missing" Flag Across Years')
plt.xlabel('Year')
plt.ylabel('Count of Missing Flags')

# Distribution of 'Male Missing' across years
plt.subplot(3, 1, 2)
sns.countplot(data=ontario_data[ontario_data['Male Missing'] == 1], x="Year", color='green')
plt.title('Distribution of "Male Missing" Flag Across Years')
plt.xlabel('Year')
plt.ylabel('Count of Missing Flags')

# Distribution of 'Female Missing' across years
plt.subplot(3, 1, 3)
sns.countplot(data=ontario_data[ontario_data['Female Missing'] == 1], x="Year", color='red')
plt.title('Distribution of "Female Missing" Flag Across Years')
plt.xlabel('Year')
plt.ylabel('Count of Missing Flags')

# Display the plots
plt.tight_layout()
plt.show()
```

The visualizations provide insights into the distribution of missing data flags across years:

Distribution of "Both sexes Missing" Flag Across Years:

The distribution is relatively consistent across years, with some fluctuations. There doesn't seem to be a specific year where the missing data flag for both sexes is significantly higher.
Distribution of "Male Missing" Flag Across Years:

Similar to the "Both sexes Missing" flag, the distribution of the "Male Missing" flag is relatively consistent across years. Some fluctuations are present, but there's no clear pattern of specific years with exceptionally high missing flags.
Distribution of "Female Missing" Flag Across Years:

The distribution pattern for the "Female Missing" flag is consistent with the other two flags. Some years have a slightly higher count, but overall, the distribution is relatively even across years.
In conclusion, while there are records with missing flags across various years, there doesn't seem to be a specific year where data is predominantly missing. The data appears to be evenly distributed, with no clear pattern of missingness based on the year.
